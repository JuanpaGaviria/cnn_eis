## With the following code, you will upload the dataset to a data storage account, the one created by default when creating the workspace in AML

```python
from azureml.core import Workspace
from azureml.core import Dataset
from azureml.data.datapath import DataPath

ws = Workspace.from_config()
datastore = ws.get_default_datastore()
Dataset.File.upload_directory(src_dir='images',
                              target=DataPath(datastore, '/dataset')
                                )
```

### The following code register the dataset as a file dataset

```python
from azureml.core import Workspace
from azureml.core import Dataset

ws = Workspace.from_config()
datastore = ws.get_default_datastore()
file_data_set = Dataset.File.from_files(path=(datastore, 'dataset'))  # Create FileDataset
# Registering the file datset
try:
    file_data_set = file_data_set.register(workspace = ws,
                                        name = 'dataset',
                                        tags = {'format' : 'JPG'},
                                        create_new_version = True)
except Exception as ex:
    print(ex)

print('Dataset registered')
```
